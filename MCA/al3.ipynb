## Matplotlib practicals-Bar Chart

- Objective: Visualize categorical data

```python
import matplotlib.pyplot as plt
```

```python
fruits = ['Apple', 'Banana', 'Cherry', 'Grapes']
quantities = [10, 15, 7, 12]
```

```python
plt.bar(fruits, quantities, color=['red', 'yellow', 'pink', 'purple'])
plt.xlabel('Fruits')
plt.ylabel('Quantities')
plt.title('Fruit Quantities')
plt.show()
```

```python
# Horizontal Bar Chart
plt.barh(fruits, quantities, color=['red', 'yellow', 'pink', 'purple'])
plt.xlabel('Quantities')
plt.ylabel('Fruits')
plt.title('Fruit Quantities')
plt.show()
```
