## Multiple Lines in One Chart

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, label="Series 1", marker='o')
plt.plot(x, y2, label="Series 2", marker='s')
plt.title("Multiple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show labels
plt.show()
```
