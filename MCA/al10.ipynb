## Bubble chart Practicals

- Step 1: Create Sample Dataset

```python
products = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [100, 200, 300, 400, 500],
    'Profit': [20, 30, 50, 10, 60],
    'MarketShare': [15, 40, 20, 10, 50]
}
```

- Step 2: Load Dataset

```python
import pandas as pd
df = pd.DataFrame(products)
print(df)
```

- Step 3: Bubble chart using Matplotlib

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(
    df['Profit'],           # x-axis
    df['Sales'],            # y-axis
    s=df['MarketShare']*10, # bubble size (scaled)
    alpha=0.6,              # transparency
    c=df['MarketShare'],    # bubble color
    cmap='coolwarm'         # colormap
)

for i, row in df.iterrows():
    plt.text(row['Profit']+0.5, row['Sales']+5, row['Product'])

plt.title('Product Sales vs Profit Bubble Chart')
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.colorbar(label='Market Share')
plt.show()
```

- Step 4: Bubble Chart Using Plotly (Interactive)

```python
# %pip install -q --upgrade "nbformat>=4.2.0"

import plotly.express as px

fig = px.scatter(
    df,
    x='Profit',
    y='Sales',
    size='MarketShare',
    color='MarketShare',
    hover_name='Product',
    size_max=60,
    title='Interactive Bubble Chart: Product Sales vs Profit'
)

fig.show()
```

- Step 5: Practical Exercises

```python
# Exercise 1: Scale bubble size differently and see how it affects the plot.

plt.scatter(
    df['Profit'],
    df['Sales'],
    s=df['MarketShare']*20,  # Changed scaling factor
    alpha=0.6,
    c=df['MarketShare'],
    cmap='coolwarm'
)
plt.title('Scaled Bubble Size')
plt.show
```

```python
# Exercise 2: Change the colormap in Matplotlib ('viridis', 'plasma', 'inferno').
plt.scatter(
    df['Profit'],
    df['Sales'],
    s=df['MarketShare']*10,
    alpha=0.6,
    c=df['MarketShare'],  # use numeric data for colors
    cmap='plasma'         # apply desired colormap
)
plt.title('Different Colormap')
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.colorbar(label='Market Share')
plt.show()
```

```python
# Exercise 3: Add another dimension using marker shape or symbol in Plotly (e.g., Product Category).
df['Category'] = ['X', 'Y', 'X', 'Y', 'X']  # New categorical dimension
fig = px.scatter(
    df,
    x='Profit',
    y='Sales',
    size='MarketShare',
    color='Category',  # Use new categorical dimension for color
    symbol='Category', # Different symbols for categories
    hover_name='Product',
    size_max=60,
    title='Bubble Chart with Categorical Dimension'
)
fig.show()
```
