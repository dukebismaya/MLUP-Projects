## Seaborn Iris visualization dashboard

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset
iris = sns.load_dataset('iris')
```

```python
sns.set_theme(style="whitegrid", palette="muted")

plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)

#pairplot for all features
sns.pairplot(iris, hue="species", height=2.5)
plt.title('Pairplot of Iris Dataset')
```

```python
# Boxplot for Sepal Length
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Boxplot of Sepal Length by Species')
```

```python
# Violin plot for Petal Length
plt.subplot(2, 2, 3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Violin Plot of Petal Length by Species')
```

```python
# Heatmap for correlation
plt.subplot(2, 2, 4)
sns.heatmap(iris.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Feature Correlations')
```

```python
# Scatter plot for Petal Length vs Petal Width
plt.figure(figsize=(8,5))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', style='species', s=100, data=iris)
plt.title('Petal Length vs Petal Width')
plt.show()
```
