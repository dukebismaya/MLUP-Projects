## Histogram practicals
- Random data histograms
- Real dataset histograms (Iris & others)
- Histograms with KDE
- Multiple groups comparison
- Normalized histograms
- Horizontal histograms
- Custom styling and colors

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris, fetch_california_housing
```

```python
# 1. Random Data Histograms
data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Histograms with Real Datasets
# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names) #type: ignore
plt.hist(iris_df['sepal length (cm)'], bins=20, alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram of Iris Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Histograms with KDE
sns.histplot(data=iris_df, x='sepal length (cm)', bins=20, kde=True, color='purple')
plt.title('Histogram with KDE of Iris Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
## Horizontal histograms
plt.hist([iris_df[iris.target == i]['sepal length (cm)'] for i in range(3)],
         bins=20, stacked=True, label=iris.target_names, edgecolor='black', orientation='horizontal')
plt.title('Horizontal Histogram of Iris Sepal Length by Species')
plt.xlabel('Frequency')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# Normalized Histogram (Probability)
plt.hist(data, bins=25, density=True, color='cyan', edgecolor='black', alpha=0.7)
plt.title('Normalized Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# 4. Multiple Groups Comparison
group1 = np.random.normal(loc=0, scale=1, size=500)
group2 = np.random.normal(loc=2, scale=1.5, size=500)
group3 = np.random.normal(loc=-2, scale=0.5, size=500)

plt.hist([group1, group2, group3], bins=30, stacked=True, label=['Group 1', 'Group 2', 'Group 3'], color=['cyan', 'magenta', 'yellow'], edgecolor='black', alpha=0.7)
plt.legend()
plt.title('Stacked Histogram of Multiple Groups')
plt.show()
```

```python
# Histogram with different bin sizes
plt.hist(data, bins=[-3, -2, -1, 0, 1, 2, 3], color='magenta', edgecolor='black')
plt.title('Histogram with Custom Bin Sizes')
plt.show()
```

```python
# Overlaying Multiple KDEs (Seaborn)
sns.histplot(group1, bins=30, kde=True, color='blue', label='Group 1')
sns.histplot(group2, bins=30, kde=True, color='orange', label='Group 2')
sns.histplot(group3, bins=30, kde=True, color='green', label='Group 3')
plt.legend()
plt.title('Overlayed KDEs for Two Groups')
plt.show()
```

```python
# Histogram of California Housing Prices
california = fetch_california_housing()
california_df = pd.DataFrame(data=california.data, columns=california.feature_names) #type: ignore
plt.hist(california_df['MedInc'], bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title('Histogram of California Median Income')
plt.xlabel('Median Income (10k USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram with Transparency
plt.hist([group1, group2], bins=30, color=['red', 'blue'], alpha=0.5, label=['Group 1', 'Group 2'], edgecolor='black')
plt.legend()
plt.title('Histogram with Transparency')
plt.show()
```

```python
# Seaborn Style Customization
sns.set_style('whitegrid')
sns.histplot(data=iris_df, x='petal length (cm)', bins=20, kde=True, color='teal')
plt.title('Seaborn Styled Histogram of Iris Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
