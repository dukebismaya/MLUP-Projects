## Bar Chart with Multiple Series (Grouped Bars)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales_2024 = [200, 220, 250, 210]
sales_2025 = [210, 230, 240, 220]

x = np.arange(len(months))  # label locations
width = 0.35  # width of bars

plt.bar(x - width/2, sales_2024, width, label='2024', color='orange')
plt.bar(x + width/2, sales_2025, width, label='2025', color='purple')

plt.xticks(x, months)
plt.ylabel("Sales")
plt.title("Monthly Sales Comparison")
plt.legend()
plt.show()
```

# Practical points:
- `width` controls the thickness of bars
- Shift bars left/right to create grouped bars
