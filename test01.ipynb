## Elbow Method helps to find the optimal number of clusters (k) in K-Means clustering

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data

# Apply KMeans with different cluster numbers
inertia = []   # Sum of squared distances (WCSS)

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title("Elbow Method for Optimal k (Iris Dataset)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (WCSS)")
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```
